```python
def plota_gastos_por_mes(dados, figsize=(10,6)):  #criar uma função reutilisável
  axis = dados.T.plot(figsize=figsize)
  axis.set_ylabel("gasto aprobado em milhões de reais")
  axis.set_xlabel("mes")
```

```python
plota_gastos_por_mes(ordenados_por_total[meses_filtrados].head(3))
```

```python
plota_gastos_por_mes(ordenados_por_total[meses_filtrados].head(3))
plt.ylim(0,600) #alterar o intervalo do eixo Y
```

```python
plota_gastos_por_mes(ordenados_por_total[ordenados_por_total.columns[47:70]].head(3)) # alterar o valor do eixo x (intervalo as colunas de 47 a 70)
plt.ylim(0,300)
```

```python
plota_gastos_por_mes(ordenados_por_total[ordenados_por_total.columns[-12:]].head(3)) # alterar o valor do eixo x (últimos 12 meses )
plt.ylim(110,450)
```

```python
plota_gastos_por_mes(ordenados_por_total[ordenados_por_total.columns[-60:]].head(5)) # alterar o valor do eixo x(últimos 60 meses )
plt.ylim(0,600)
```

```python
plota_gastos_por_mes(ordenados_por_total[ordenados_por_total.columns[-60:]].head(5), figsize=(6,6)) # alterar o valor do eixo x(últimos 60 meses )
plt.ylim(0,600)
```
